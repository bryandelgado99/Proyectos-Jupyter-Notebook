```python
#Se creara un método que procese por lotes cada uno de los archivos .txt que componen la constitucion
#Primer archivo
archi = open('constitucion/1.txt', 'r')
text1 = archi.read()

#Creamos una variable que convierta en lista las palabras
diccionario = text1.split()

#Creamos una lista para la frecuencia
frecuencia = []

for x in text1:
    frecuencia.append(text1.count(x))

#Creamos un diccionario con ambas listas
glosario = (dict(zip(diccionario, frecuencia)))
```
