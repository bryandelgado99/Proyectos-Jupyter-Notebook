```python
#Se creara un método que procese por lotes cada uno de los archivos .txt que componen la constitucion
#Primer archivo
archi = open('constitucion/1.txt', 'r')
text1 = archi.read()

#Creamos una variable que convierta en lista las palabras
diccionario = text1.split()

#Creamos una lista para la frecuencia
frecuencia = []

for x in text1:
    frecuencia.append(text1.count(x))

#Creamos un diccionario con ambas listas
print("El diccionario es: \n")
glosario = (dict(zip(text1, frecuencia)))

#Ordenamos el diccionario
glosario_items = glosario.items()

for item in glosario_items:
    print(f'{item[0]}: {item[1]}')

#Ordenamos el diccionario decendente por numero de repeticions
glosario_sorted = dict(sorted(glosario.items(), key=lambda item:item[1], reverse=True))
glosario_sorted
```
